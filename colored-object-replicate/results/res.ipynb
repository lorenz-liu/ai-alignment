```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('cobjs_path_patching/cfg_cobjs_context_movers_gpt2med.npy')


average_effect = np.mean(data)

print(f'Average Effect: {average_effect}')

plt.figure(figsize=(10, 8))
plt.title('Effect of Path Patching Across Layers and Heads')
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar(label='Effect Magnitude')
plt.xlabel('Heads')
plt.ylabel('Layers')
plt.show()

```
